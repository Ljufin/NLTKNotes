Calculate some statistics for some texts.

```python
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
    print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocab), fileid)
```

This output displays the average number of characters in a word, the average number of words in a sentence, and the average number of times that a vocabulary word appears in the text for each text in our collection. 

The word length in English seems to be 4-5 words and the other two statistics depend on the author, and maybe the genre as well(plays have really short sentences that bring down the average).

# 1.3  Brown Corpus
The Brown corpus was created at Brown University in 1961 and is the first million word corpus. The texts are sourced from a bunch of different categories.

```python
from nltk.corpus import brown
print(brown.categories())
print(brown.words(categories='news'))
print(brown.words(fileids=['cg22']))
print(brown.sents(categories=['news', 'editorial', 'reviews']))
```

Because of inclusion of multiple genres, the Brown corpus is the perfect place to study the differences between genres. This is called *stylistics*.

```python
# this code will print the distribution of certain words in news texts
news_text = brown.words(categories='news')
fdist = nltk.FreqDist(w.lower() for w in news_text)
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
    print(m + ':', fdist[m], end=' ')
```

```python
# counting the number of wh words in religion texts
religion_text = brown.words(categories='news')
fdist = nltk.FreqDist(w.lower() for w in religion_text)
wh_words = ['what', 'when', 'where', 'who', 'why']
for w in wh_words:
    print(w + ':', fdist[w], end=' ')
```

We can compare multiple different genres. Notice that hobbies involve a lob of modals.

```python
cdf = nltk.ConditionalFreqDist( 
    (genre, word) 
    for genre in brown.categories() 
    for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
# a nice way of arranging the data in a table
cdf.tabulate(conditions=genres, samples=modals)
```

# 1.5   Inaugural Address Corpus
This is a collection of all the inaugrual addresses for the U.S. presidents There are 55 texts in total. A nice feature of this corpus is that there is a time data point hidden in each of the file names, which can easily be extracted.

```python
from nltk.corpus import inaugural
print([fileid[:4] for fileid in inaugural.fileids()][-10:])
```

Using this time variable, we can compare differing usages of words over the years.

```python
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:4])
    for fileid in inaugural.fileids()
    for w in inaugural.words(fileid)
    for target in ['america', 'citizen']
    if w.lower().startswith(target)) 
cfd.plot()
```

The years are kinda squished at the bottom, but it looks like the word "america" is getting more popular in recent years. Let's compare a few more words and make the graph really cluttered:

```python
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:4])
    for fileid in inaugural.fileids()
    for w in inaugural.words(fileid)
    for target in ['united', 'freedom', 'patriot', 'slavery', 'god']
    if w.lower().startswith(target)) 
cfd.plot()
```
